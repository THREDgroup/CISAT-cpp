```python
%matplotlib inline
import scipy.stats as st
```

# Load the Results and Parameters

```python
results = loadtxt('../test/output_bench.txt')
```

```python
with open('../test/example_bench.txt') as f:
    for line in f:
        line = line.split()
        if line[0] == "NAGENTS":
            n_agents = float(line[1])
        if line[0] == "NREPS":
            n_reps = int(line[1])
```

# Plot Performance v. Time

```python
figsize(8, 8)
x = [y*n_agents for y in range(size(results, axis=0))]
xlabel("Total Objective Function Evaluations")
ylabel("Best Solution Found")
h1 = plot(x, results, 'gray', lw=0.1)
h2, = plot(x, mean(results, axis=1))
h3, = plot(x, st.gmean(results, axis=1))
legend([h1[0], h2, h3], ["Raw Data", "Arithmetic Mean", "Geometric Mean"]);
```

# Plot Cumulative Distribution Functions

```python
y = [float(y)/n_reps for y in range(n_reps)]
xlabel("Objective Function Value")
ylabel("CDF")
h1, = plot(sort(results[-1]), y)
h2, = plot(sort(results[0]), y)
legend([h1, h2], ["Final Distribution", "Initial Distribution"])
```
