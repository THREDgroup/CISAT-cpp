```python
%pylab inline
import scipy.stats as st
```

# Load the Results and Parameters

```python
results_bench = loadtxt('../test/output_bench.txt')
results_metabench = loadtxt('../test/output_metabench.txt')
```

```python
with open('../test/example_bench.txt') as f:
    for line in f:
        line = line.split()
        if line[0] == "NAGENTS":
            n_agents_bench = float(line[1])
        if line[0] == "NREPS":
            n_reps_bench = int(line[1])
            
with open('../test/output_meta.txt') as f:
    for line in f:
        line = line.split()
        if line[0] == "NAGENTS":
            n_agents_metabench = float(line[1])
        if line[0] == "NREPS":
            n_reps_metabench = int(line[1])
```

# Plot Performance v. Time

```python
figsize(16, 8)

subplot(121)
title("Before Meta-Optimization")
x = [y*n_agents_bench for y in range(size(results_bench, axis=0))]
xlabel("Total Objective Function Evaluations")
ylabel("Best Solution Found")
h1 = plot(x, results_bench, 'gray', lw=0.1)
h2, = plot(x, mean(results_bench, axis=1))
h3, = plot(x, st.gmean(results_bench, axis=1))
grid('on')
legend([h1[0], h2, h3], ["Raw Data", "Arithmetic Mean", "Geometric Mean"]);
yscale('log')

subplot(122)
title("After Meta-Optimization")
x = [y*n_agents_metabench for y in range(size(results_metabench, axis=0))]
xlabel("Total Objective Function Evaluations")
ylabel("Best Solution Found")
h1 = plot(x, results_metabench, 'gray', lw=0.1)
h2, = plot(x, mean(results_metabench, axis=1))
h3, = plot(x, st.gmean(results_metabench, axis=1))
grid('on')
legend([h1[0], h2, h3], ["Raw Data", "Arithmetic Mean", "Geometric Mean"]);
yscale('log')

tight_layout()
```

# Plot Cumulative Distribution Functions

```python
figsize(12, 8)

subplot(211)
title("Before Meta-Optimization")
y = [float(y)/n_reps_bench for y in range(n_reps_bench)]
xlabel("Objective Function Value")
ylabel("CDF")
h1, = plot(sort(results_bench[-1]), y)
h2, = plot(sort(results_bench[0]), y)
legend([h1, h2], ["Final Distribution", "Initial Distribution"], 2)
# xlim(0, 16)
xscale('log')
grid('on')

subplot(212)
title("Before Meta-Optimization")
y = [float(y)/n_reps_metabench for y in range(n_reps_metabench)]
xlabel("Objective Function Value")
ylabel("CDF")
h1, = plot(sort(results_metabench[-1]), y)
h2, = plot(sort(results_metabench[0]), y)
legend([h1, h2], ["Final Distribution", "Initial Distribution"], 2);
# xlim(0, 16)
xscale('log')
grid('on')

tight_layout()
```
